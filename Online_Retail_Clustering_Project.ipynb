# **Project Summary -**

The Online Retail dataset is a popular transactional dataset that captures a full year’s worth of online purchases made between December 2010 and December 2011. It originates from a UK-based online store and contains over 500,000 rows, making it suitable for a wide range of data analysis and machine learning tasks. Each row in the dataset represents a transaction or invoice and includes several important columns such as `InvoiceNo`, `StockCode`, `Description`, `Quantity`, `InvoiceDate`, `UnitPrice`, `CustomerID`, and `Country`. These fields provide a detailed view of the products purchased, their quantities, unit prices, and the countries from which the orders were placed.

The dataset’s primary value lies in its richness and realism. Since it includes real transaction data, it reflects real-world customer behavior, seasonality, and purchasing patterns. The majority of transactions originate from the United Kingdom, which is expected as the business is based there. However, there is also notable activity from countries like Germany, France, the Netherlands, and Spain, offering opportunities for cross-country comparative analysis. This geographic information can help businesses identify their most valuable customer bases and prioritize marketing or operational efforts accordingly.

Despite its usefulness, the dataset contains several challenges that require careful preprocessing. Many entries have missing `CustomerID` values, which can limit the scope of customer-level analyses. There are also duplicate records and canceled transactions, identified by invoice numbers beginning with the letter "C", which must be filtered out to avoid skewed results. Additionally, some rows contain negative or zero quantities and prices, often indicating returns or entry errors. These anomalies should be handled or removed based on the intended analysis.

Once cleaned, the dataset becomes a powerful tool for various analytical purposes. One common use case is customer segmentation using clustering algorithms like K-Means or Hierarchical Clustering. By grouping customers based on features such as purchase frequency, monetary value, and product types, businesses can tailor their marketing strategies to different customer segments. Another valuable application is building a recommendation system, where past purchase data can be used to suggest new products to customers using collaborative filtering or content-based approaches.

The dataset is also well-suited for market basket analysis, where algorithms like Apriori or FP-Growth can reveal associations between frequently bought products. These insights help in designing effective cross-selling strategies and promotional offers. Furthermore, analyzing trends over time using the `InvoiceDate` column allows businesses to understand seasonality, track peak periods, and forecast future demand.

In conclusion, the Online Retail dataset is a comprehensive and realistic resource for learning and applying data science techniques in an e-commerce context. While it requires thoughtful cleaning and preprocessing, its rich features make it suitable for tasks like clustering, recommendation, and trend analysis. It offers deep insights into customer purchasing behavior, helping businesses make data-driven decisions and improve overall customer engagement.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
df=df.copy()
df.head()
````

````python
# Converting the 'InvoiceDate' column to datetime format
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'],errors='coerce',format='%m/%d/%y %H:%M')
````

````python
# Extracting month name,week name,day,hour,year,min from the 'InvoiceDate' column
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
df['InvoiceWeek']  = df['InvoiceDate'].dt.isocalendar().week
df['InvoiceDay']   = df['InvoiceDate'].dt.day
df['InvoiceHour']  = df['InvoiceDate'].dt.hour
df['InvoiceYear']  = df['InvoiceDate'].dt.year
df['InvoiceMin']   = df['InvoiceDate'].dt.minute
````

````python
# Creating new column 'TotalSpend' by multiplying quantity with unitprice
df['TotalSpend'] = df['Quantity'] * df['UnitPrice']
````

````python
# Replace missing CustomerID values with 0.0 (indicating unknown customers)
df['CustomerID'] = df['CustomerID'].fillna(0.0)
````

````python
# Drop all rows where the 'Description' field is missing (NaN)
df = df.dropna(subset=['Description'])
````

````python
# Remove duplicate rows from the dataset, keeping only the first occurrence
df = df.drop_duplicates(keep='first')
````

````python
# Handling Outliers & Outlier treatments
import seaborn as sns
import matplotlib.pyplot as plt

# List of columns to check
columns = ['Quantity', 'UnitPrice', 'CustomerID', 'InvoiceMonth', 'InvoiceWeek','InvoiceDay', 'InvoiceHour', 'InvoiceYear', 'InvoiceMin']

# Function to check outliers using seaborn boxplots
def check_outliers(columns, data):
    for col in columns:
        plt.figure(figsize=(6, 5))
        sns.boxplot(y=data[col], color='skyblue')
        plt.title(f'Boxplot of {col}')
        plt.tight_layout()
        plt.show()

# Example usage:
check_outliers(columns, df)
````

````python
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower=lower_bound, upper=upper_bound)
````

````python
df['UnitPrice'] = cap_outliers(df['UnitPrice'])
df['Quantity'] = cap_outliers(df['Quantity'])
````

## ***7. Cluster Model Implementation***

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Take only the first 10,000 rows for clustering
X_sample = X[:10000]

wcss = []
for i in range(1, 22):
    kmeans = KMeans(
        n_clusters=i,
        init='k-means++',
        max_iter=300,
        n_init=10,
        random_state=0
    )
    kmeans.fit(X_sample)
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8,5))
plt.plot(range(1, 22), wcss)
plt.title('Elbow Method (on 10,000 samples)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
````


We used the **elbow method** to decide the number of clusters.

In the graph, the WCSS (error) goes down as we increase clusters, but after **5 clusters** the curve starts to flatten.

This means adding more clusters after 5 does not improve the results much.
So, we choose **5 clusters** as the best and most balanced option.


````python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Take only the first 10,000 rows to make it feasible
X_sample = X[:10000]

# Compute linkage on this subset
linkage_matrix = sch.linkage(X_sample, method='ward')

# Plot dendrogram
plt.figure(figsize=(13,8))
sch.dendrogram(linkage_matrix)
plt.title('Dendrogram (on 10,000 samples)')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
plt.show()
````

This is a dendrogram from hierarchical clustering.

We look for a big jump in the height of the lines.

At that point (around distance 6–7), the data naturally splits into about **5 clusters**, so we chose 5.


### Agglomerative Clustering

````python
from sklearn.cluster import AgglomerativeClustering

# Take only the first 10,000 rows
X_sample = X[:10000]

#  Apply Agglomerative Clustering
hc = AgglomerativeClustering(
    n_clusters=5,          # you can change this based on your dendrogram
    metric='euclidean',   # distance metric
    linkage='ward'          # linkage method
)

y_hc = hc.fit_predict(X_sample)
````

````python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Take only first 10,000 rows for efficiency
X_sample = X[:10000]

silhouette_scores = []  # to store [n_clusters, score]
range_n_clusters = range(2, 8)  # clusters from 2 to 15

for n_clusters in range_n_clusters:
    clusterer = AgglomerativeClustering(
        n_clusters=n_clusters,
        metric='euclidean',   # ✅ updated parameter
        linkage='ward'
    )
    preds = clusterer.fit_predict(X_sample)

    score = silhouette_score(X_sample, preds)
    silhouette_scores.append([n_clusters, round(score, 3)])
    print(f"For n_clusters = {n_clusters}, silhouette score = {score:.3f}")
````

After seeing the silhoutte scores since the cluster 4 has hight score then others so it is being considered
